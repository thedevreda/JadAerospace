```python
# Libs
import pandas as pd
import os
```

```python
pwd = os.getcwd()
# Dataset
dataset = pd.read_csv(pwd + r"\Scraping\Data - Jadaero Parts.csv")
dataset = dataset.copy()
dataset.info()
```

`DataSet : partNumber | partName | description`

- Part name = Partnumber + PartName
- Description = PartNumber + PartName

```python
dataset.sample()
```

```python
dataset.duplicated().sum()

# NO DUBLICATES
```

```python
dataset[['PartNumber', 'PartName']] = dataset['Part Name'].str.split(' ', n=1, expand=True)
dataset

# SPLITTING PART NAME COLUMN TO PARTNUMBER AND PARTNAME
```

```python
dataset.drop(columns=['Quantity','Part Name'], inplace=True)
dataset.sample(5)

# REMOVE COLUMNS 
```

```python
dataset[dataset['PartNumber'] == ''].value_counts()
dataset.info()
# CHECKING FOR EMPTY PARTNUMBER
```

```python
dataset = dataset[['PartNumber', 'PartName', 'Description']]  # New order
dataset.sample(5)


#ORDERING THE COLUMNS TO NEW ORDER
```

```python
dataset[dataset["PartName"] == 'VALVE']
```

```python
dataset['PartName'].unique().tolist() 
# 'TYRE MAIN', 'TIRE – MAIN', FOR SALE / TRIM ESPACE - Viz Damag
```

```python
dataset['PartName'] = dataset['PartName'].str.replace('TYRE', 'TIRE', case=False, regex=False)

dataset['PartName'] = dataset['PartName'].str.strip('– ')  # Trimming spaces and dashes

dataset['PartName'].unique().tolist() 



# CORRECTING THE SPELLING AND TRIMING ESPACES AND DASHES
```

```python
dataset[dataset['PartName'] == 'For Sale']

dataset['PartName'] = dataset['PartName'].str.replace('For Sale', 'ENGINE NEW 675 HP')

dataset[dataset['PartName'] == 'ENGINE NEW 675 HP']

# REPLACING 'For Sale' WITH 'ENGINE NEW 675 HP'
```

```python
dataset.sample(5)
```

```python
dataset.to_csv(pwd + r"\Data\Data - Jadaero Parts Cleaned.csv", index=False)

# SAVING THE CLEANED DATASET
```
