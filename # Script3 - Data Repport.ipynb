```python
# Libs
import pandas as pd
import os
```

```python
pwd = os.getcwd()
# Dataset
dataset = pd.read_csv(pwd + r"\Data\Data - Jadaero Parts Edited.csv")
dataset = dataset.copy()
dataset.info()
```

```python
dataset['Risk Level'].unique()
```

```python
# Client Type 1: Regional Airlines (Risk-Averse, Cost-Conscious)
# Risk Level: 'Low', 'Moderate' i wont filter this out cuz i'll filter in excel from the extremw to low after extracting the data
# Replacement Interval: Longer intervals like '10 years', '15 years', '7 years'
# Geo Targets: 'High-frequency routes (US domestic, Europe-Middle East)'
# Boeing/Bombardier Targets: Focus on common regional aircraft
#

clt1 = dataset[
        dataset['Replacement Interval'].isin(['10 years', '15 years', '7 years']) |
        dataset['Geographical Targets'].str.contains('High-frequency routes') |
        dataset['Bombardier Targets'].str.contains('Regional aircraft') |
        dataset['Boeing Targets'].str.contains('Regional aircraft')]

clt1.sample(2)
```

```python
# Client Type 2: Cargo Airlines (Performance-Focused)
# Risk Level: 'High', 'Extreme' (willing to pay for performance)
# Replacement Interval: '5,000hrs' intervals, 'On-condition' monitoring
# Geo Targets: 'Harsh environments (desert, coastal) + long-haul operations', 'Hot climates + cargo temperature-sensitive operations'
# MRO Targets: Heavy maintenance specialists

Geo = 'Harsh environments|(desert, coastal)|long-haul operations|Hot climates|cargo temperature-sensitive operations'
clt2 = dataset[
        dataset['Replacement Interval'].isin(['5,000hrs']) |
        dataset['Geographical Targets'].str.contains(Geo) |
        dataset['MRO Targets'].str.contains('Heavy maintenance')]
clt2.sample(2)
```

```python
# Client Type 3: MRO Service Providers
# Risk Level: All levels (they service diverse fleets)
# Replacement Interval: Focus on frequently replaced items like '6 months', '2,000hrs', '500 cycles'
# Geo Targets: 'Safety-critical universal demand + harsh environments'
# MRO Targets: Primary focus
Geo2 = 'Safety-critical universal demand|harsh environments'
clt3 = dataset[
        dataset['Replacement Interval'].isin(['6 months', '2,000hrs', '500 cycles']) |
        dataset['Geographical Targets'].str.contains(Geo2) |
        dataset['MRO Targets'].str.contains('Primary focus')]
clt3.sample(2)
```

```python
# Client Type 4: Legacy Fleet Operators
# Risk Level: 'Moderate', 'High'
# Replacement Interval: 'On-condition' with monitoring meaning Uknown that's why i wont filter this out
# Geo Targets: 'Older fleets needing upgrades + EM interference environments'
# Boeing Targets: Older Boeing models

Geo3 = 'Older fleets needing upgrades|EM interference environments'
clt4 = dataset[
        dataset['Geographical Targets'].str.contains(Geo3) |
        dataset['Boeing Targets'].str.contains('Older Boeing models')]
clt4.sample(2)
```

```python
# Export both to one Excel file, in different sheets
with pd.ExcelWriter(pwd + '\Data\Data - Final Rapport Output.xlsx', engine='openpyxl') as writer:
    clt1.to_excel(writer, sheet_name='Regional Airlines ', index=False)
    clt2.to_excel(writer, sheet_name='Cargo Airlines', index=False)
    clt3.to_excel(writer, sheet_name='MRO Service Providers', index=False)
    clt4.to_excel(writer, sheet_name='Legacy Fleet Operators', index=False)
```
